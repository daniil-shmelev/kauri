```python
import kauri as kr
```

## Next tree with respect to lexicographic order

```python
t = kr.Tree([[[[]],[]],[[]]])
kr.display(t)
print("Next tree:")
kr.display(next(t))
```

## All trees of a given order

```python
for n in range(6):
    print("Trees of order " + str(n) + ":")
    for t in kr.trees_of_order(n):
        kr.display(t)
```

## All trees up to a given order

```python
print("All trees up to order 4:")
for t in kr.trees_up_to_order(4):
    kr.display(t)
```
