# Table 1: $\psi$ = Implicit midpoint rule

Print the following:

$\tau$<br>
$(-1)^{|\tau|}S\tau$<br>
$1/\tau!$<br>
$\psi(\tau)$<br>
$(-1)^{|\tau|}\psi(S\tau)$<br>

```python
psi = kr.implicit_midpoint.elementary_weights_map()
m = kr.sign & bck.antipode

for t in kr.trees_up_to_order(4):
    kr.display(t)
    s = m(t).singleton_reduced()
    kr.display(s)
    print(1 / t.factorial(), psi(t), psi(s))
    print(50*"#")
```
