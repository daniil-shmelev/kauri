# Linear Stability Domains of EES Schemes
kauri can be used to verify the linear stability domains of Runge--Kutta schemes, by considering the coefficients of the B-series expansion for all ladder trees. We give an example below for EES schemes, using sympy to get exact symbolic expressions for the coefficients.

```python
def get_stability_polynomial(rk_a, rk_b, s, max_order = 5):
    z = sympy.symbols('z')
    a = sympy.Matrix(s, s, lambda i, j: sympy.symbols(f'a{i}{j}'))
    b = sympy.Matrix(1, s, lambda i, j: sympy.symbols(f'b{j}'))

    expression = 1

    z_pow = 1
    t = kr.Tree(None)
    for i in range(max_order):
        t = t.as_forest().join() # Grow the ladder
        z_pow *= z
        r = kr.rk_symbolic_weight(t, s) # Get the elementary weight as a sympy expression

        # Substitute RK coefficients symbolically
        for i in range(s):
            for j in range(s):
                r = r.subs(a[i,j], rk_a[i,j])
            r = r.subs(b[i], rk_b[i])
            
        expression += r * z_pow # Update polynomial
    return sympy.Poly(expression, z).as_expr()
```

## $EES(2,5;x)$

**Theorem:** The linear stability domain of an $\text{EES}(2,5;x)$ scheme is given by
$$ \{z \in \mathbb{C} : |1 + z + \frac{1}{2}z^2 + \frac{1}{8}z^3| < 1\} $$

```python
ees25_a = sympy.Matrix(
    [
        [0, 0, 0],
        [sympy.Rational(1,3), 0, 0],
        [sympy.Rational(-5,48), sympy.Rational(15,16), 0]
    ]
)

ees25_b = sympy.Matrix([sympy.Rational(1,10), sympy.Rational(1,2), sympy.Rational(2,5)])
```

```python
get_stability_polynomial(ees25_a, ees25_b, 3)
```

## $EES(2,7;x)$

**Theorem:** The linear stability domain of an $\text{EES}(2,7;x)$ scheme is given by
$$ \{z \in \mathbb{C} : |1 + z + \frac{1}{2}z^2 + \frac{1}{4}(2 - \sqrt{2})z^3 + \frac{1}{8}(3 - 2\sqrt{2})z^4| < 1\} $$

```python
ees27_a = sympy.Matrix(
    [
        [0, 0, 0, 0],
        [sympy.Rational(1,3) * (2 - sympy.sqrt(2)), 0, 0, 0],
        [sympy.Rational(1,24) * (-4 + sympy.sqrt(2)), sympy.Rational(1,8) * (4 + sympy.sqrt(2)), 0, 0],
        [sympy.Rational(1,168) * (-176 + 145* sympy.sqrt(2)), sympy.Rational(3,56) * (8 - 5*sympy.sqrt(2)), sympy.Rational(3,7) * (3 - sympy.sqrt(2)), 0]
    ]
)

ees27_b = sympy.Matrix([sympy.Rational(1,14) * (5 - 3*sympy.sqrt(2)), sympy.Rational(1,14) * (3 + sympy.sqrt(2)), sympy.Rational(3,14) * (-1 +2* sympy.sqrt(2)), sympy.Rational(1,14) * (9 - 4*sympy.sqrt(2))])
```

```python
get_stability_polynomial(ees27_a, ees27_b, 4)
```
