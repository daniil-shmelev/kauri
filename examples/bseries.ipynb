```python
import kauri as kr
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

# Truncated B-Series Approximations

```python
y1 = sp.symbols('y1')
y = sp.Matrix([y1])

f = sp.Matrix([y1 ** 2])

m = kr.exact_weights
bs2 = kr.BSeries(y, f, m, 2)
bs4 = kr.BSeries(y, f, m, 6)
bs6 = kr.BSeries(y, f, m, 10)

t = np.linspace(0, 0.9, 100)
true = [1 / (1 - x) for x in t]
plt.plot(t, true, linestyle="--", color="black")

plt.plot(t, [bs2([1], h) for h in t], color = 'firebrick')
plt.plot(t, [bs4([1], h) for h in t], color = 'mediumblue')
plt.plot(t, [bs6([1], h) for h in t], color = 'forestgreen')

plt.legend(['Exact', 'n = 2', 'n = 6', 'n = 10'], title = "Truncation order")
plt.title("B-Series approximations")

plt.show()
```

```python
print(bs4.series())
print(bs4.series().subs(bs6.y[0], 1))
```
